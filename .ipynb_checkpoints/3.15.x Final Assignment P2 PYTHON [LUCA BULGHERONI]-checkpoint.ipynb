````python
# cycle through all the elements in the directory and, at each cycle, load 
# them to a temporary DataFrame 'tmp_df' and append them to an empty 
# DataFrame 'fnd' 
fnd = pd.DataFrame([])
for file in files:
    if file != '.DS_Store':
        df_tmp = pd.read_excel('data/data2/' + file)
        df_tmp['Year'] = file
        #fnd = fnd.append(df_tmp)
        fnd = pd.concat([fnd, df_tmp])
fnd.head(5)
````

````python
fnd.index = pd.RangeIndex(len(fnd.index))
````

````python
fnd.shape
````

````python
#trasformo la colonna Year in intero
fnd['Year'] = fnd['Year'].astype('str').str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)
fnd.head()
````

````python
sps = pd.read_excel("data/Software Professionals Salary.xlsx")
sps
````

````python
#4A
sps_loc = sps.groupby('Location', as_index=False)[['Rating', 'Salary']].mean().round(2)
sps_loc
````

````python
sps_loc.shape
````

````python
#4B
fnd_year2021 = fnd[fnd['Year']==2021]
fnd_year2021
````

````python
fnd_loc = fnd_year2021.groupby ('HeadQuarter', as_index=False).agg({'Company/Brand':'size', 'Amount($)':'sum'})
fnd_loc
````

````python
fnd_loc.shape
````

````python
#4C
sps_fnd_loc = sps_loc.merge(fnd_loc, how='inner', left_on='Location', right_on='HeadQuarter')
sps_fnd_loc
````

````python
sps_fnd_loc.shape
````

````python
#4.D.a
sps_fnd_loc = sps_fnd_loc.drop('HeadQuarter', axis=1)
sps_fnd_loc
````

````python
#4.D.b
sps_fnd_loc['Amount($MM)'] = sps_fnd_loc['Amount($)'] / 1000000
sps_fnd_loc
````

````python
#4.D.c
sps_fnd_loc = sps_fnd_loc.drop('Amount($)', axis=1)
sps_fnd_loc
````

````python
#4.D.d
sps_fnd_loc = sps_fnd_loc.rename(columns={'Location':'City', 
                                         'Rating':'Avg. Rating', 
                                         'Salary':'Avg. Salary', 
                                         'Company/Brand':'Nr. Companies Funded', 
                                         'Amount($MM)':'Sum Funding ($MM)'})
````

### Un modo per eseguire il ciclo for selezionando solo i valori che ti interessano può essere il seguente

````python
results = soup.find("tbody")
cities = results.find_all("tr")

df_cities = pd.DataFrame([], columns=['RankPop','City','State','Population','Metropolitan','Sexratio','Literacy'])
for rows in cities.find_all("tr"): 
    row = [el.text for el in rows.find_all("td")]
    row_loc = int(row[0])-1
    df_cities.loc[row_loc] = row
````
